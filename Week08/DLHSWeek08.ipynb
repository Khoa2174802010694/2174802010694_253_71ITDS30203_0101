**8. Phân tích cảm xúc người dùng**

```python
#download các movie review
nltk.download('movie_reviews')
```

```python
import random
reviews = nltk.corpus.movie_reviews
documents = [(list(reviews.words(fileid)), category)
for category in reviews.categories()
for fileid in reviews.fileids(category)]
random.shuffle(documents)
```

```python
#Xem nội dung review đầu tiên (dòng 0, cột 0)
first_review = ' '.join(documents[0][0])
print(first_review)
```

```python
#Xem kết quả review đầu tiên (dòng 0, cột 1)
documents[0][1]
```

```python
#tạo bảng phân phối tần số các từ trong copus, bảng này cần chuyển sang dạng list
all_words = nltk.FreqDist(w.lower() for w in reviews.words())
word_features = list(all_words)
```

```python
#xác định một hàm để tính toán các đặc trưng, tức là những từ đủ quan
#trọng để thiết lập ý kiến của một review.
def document_features(document, word_features):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['{}'.format(word)] = (word in document_words)
    return features
```

```python
#tạo 1 tập các documents
featuresets = [(document_features(d,word_features), c) for (d,c) in documents]
len(featuresets)
```

```python
#tập train và tập test: 1500 dòng đầu dùng cho tập train và 500 dòng còn lại dùng cho tập test
#để đánh giá độ chính xác của mô hình.
train_set, test_set = featuresets[1500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
#tính toán độ chính xác của thuật toán
train_set, est_set = featuresets[1500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier, test_set))
```

```python
#4
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

text = """
Natural language processing allows computers to understand
and process human language in useful ways.
"""

tokens = word_tokenize(text)

stop_words = set(stopwords.words("english"))

filtered = [
    w for w in tokens
    if w.lower() not in stop_words
    and w not in string.punctuation
]

print(filtered)
```

```python
#5
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

text = "This movie is not bad but it is not very interesting."

tokens = word_tokenize(text)

stop_words = set(stopwords.words("english"))

keep = {"not", "no", "nor"}

custom_stop = stop_words - keep

filtered = [
    w for w in tokens
    if w.lower() not in custom_stop
    and w not in string.punctuation
]

print(filtered)
```

```python
#6
from nltk.corpus import wordnet as wn

word = "computer"

for synset in wn.synsets(word):
    print("Tên:", synset.name())
    print("Định nghĩa:", synset.definition())

    if synset.examples():
        print("Ví dụ:")
        for ex in synset.examples():
            print("-", ex)

    print()
```

```python
#7
from nltk.corpus import wordnet as wn

word = "good"

synonyms = set()
antonyms = set()

for synset in wn.synsets(word):
    for lemma in synset.lemmas():
        synonyms.add(lemma.name())

        for ant in lemma.antonyms():
            antonyms.add(ant.name())

print("Synonyms:")
print(sorted(synonyms))

print("\nAntonyms:")
print(sorted(antonyms))
```
